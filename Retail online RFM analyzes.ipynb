## Define the outliers


```python
for feature in ['Quantity','Price','Total_Price']:
    Q1=df[feature].quantile(0.01)
    Q3=df[feature].quantile(0.99)
    IQR=Q3-Q1
    upper=Q3+1.5*IQR
    lower=Q1-1.5*IQR
    if df[(df[feature]>upper) | (df[feature]<lower)].any(axis=None):
        print(feature,'yes')
        print(df[(df[feature]>upper) | (df[feature]<lower)].shape[0])
    else :
        print(feature,'no')
```

*Monetary: Total Customer spending

### Monetary

```python
df.head()
```

```python
df.groupby(['Customer ID']).agg({'Total_Price':'sum'}).sort_values('Total_Price',ascending=False).head()
```

```python
df_monetary=df.groupby(['Customer ID']).agg({'Total_Price':'sum'})
df_monetary.head()
```

```python
df_monetary.rename(columns={'Total_Price':'Monetary'}, inplace=True)
df_monetary.head()
```

```python
print(df_freq.shape,temp_df.shape,df_monetary.shape)
```

```python
rfm=pd.concat([df_freq,temp_df,df_monetary],axis=1)
rfm.head()
```

```python
rfm['Recency_score']=pd.qcut(rfm['Recency'],5,labels=[5,4,3,2,1])
rfm.head()
```

```python
#Facin an error due to lower frequency range/Frequency degisim araligi kucuk oldugu icin hata verdi.
rfm['Frequency_score']=pd.qcut(rfm['Frequency'],5,labels=[1,2,3,4,5])
rfm.head()
```

```python
#Thanks to Rank() method the error is amended/hatayi rank metodu ile duzeltelim
rfm['Frequency_score']=pd.qcut(rfm['Frequency'].rank(method='first'),5,labels=[1,2,3,4,5])
rfm.head()
```

```python
rfm['Monetary_score']=pd.qcut(rfm['Monetary'],5,labels=[1,2,3,4,5])
rfm.head()
```

```python
(rfm['Recency_score'].astype(str)+rfm['Frequency_score'].astype(str)+rfm['Monetary_score'].astype(str)).head()
```

```python
rfm['rfm_score']=(rfm['Recency_score'].astype(str)+rfm['Frequency_score'].astype(str)+rfm['Monetary_score'].astype(str))
rfm.head()
```

```python
#Champion group
rfm[rfm['rfm_score']=='555'].head()
```

```python
rfm[rfm['rfm_score']=='111'].head() #Hibernating
```

```python
rfm.describe().T
```

```python
#Using Regular Expression RFM map is consisted of/regular expression (duzenli ifadeler) kullanilarak RFM haritasi cikarildi.
seg_map={
    r'[1-2][1-2]' : 'Hibernating',
    r'[1-2][3-4]' : 'At Risk',
    r'[1-2]5' : 'Can\t loose',
    r'3[1-2]' : 'About to sleep',
    r'33' : "Need attention",
    r'[3-4][4-5]' : 'Loyal Customers',
    r'41' : 'Promising',
    r'51' : "New Customers",
    r'[4-5][2-3]' :'potential loyalists',
    r'5[4-5]' : 'Champions'
        
}
```

```python
rfm['segment']=rfm['Recency_score'].astype(str) + rfm['Frequency_score'].astype(str)
rfm.head()
```

```python
rfm['segment']=rfm['segment'].replace(seg_map, regex=True)
```

```python
rfm.head()
```

```python
rfm[['segment','Recency','Frequency','Monetary']].groupby('segment').agg(['mean','count']).head()
```

```python
rfm[rfm['segment']=='Loyal Customers'].head()
```

```python
rfm[rfm['segment']=='Loyal Customers'].index
```

```python
new_df=pd.DataFrame()
```

```python
new_df['Loyal Customer ID']=rfm[rfm['segment']=='Loyal Customers'].index
new_df.head()
```
